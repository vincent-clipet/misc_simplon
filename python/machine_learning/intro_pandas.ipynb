##### Import & load dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_excel("datasets/titanic.xls")
dataset.head()
```

##### Clear dataset from useless data

```python
dataset = dataset.drop(columns=['name', 'sibsp', 'parch', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest'])
dataset.head()
```

##### Print a bunch of info about the dataset (max, min, avg, median, ...)

```python
dataset.describe()
```

##### Deal with missing data

```python
# Fill with the average
# dataset.fillna(dataset['age'].mean())

# Drop lines with missing data
dataset.dropna(axis=0)

dataset.describe()
```

##### Aggregates values for a specific column

```python
dataset['pclass'].value_counts()
dataset['pclass'].value_counts().plot.bar()
```

```python
dataset["age"].hist()
```

##### GROUP BY

```python
dataset.head()
dataset.groupby(["sex"]).mean()
```

##### Change the index (0,1,2,3,...) of the dataset

```python
dataset = dataset.set_index("name")
dataset["age"]
```

##### Can use `np.array` operations on datasets too

```python
# will return an array "index|bool"
dataset["age"] > 50
```

```python
# will return the dataset lines matching the given condition
dataset[dataset["age"] > 50]
```

```python
# Will filter lines where age > 50, and then group values by their 'pclass'
dataset[dataset["age"] > 50]['pclass'].value_counts()
```
