```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
```

##### Basic simple graph

```python
x = np.linspace(0, 10, 11) # creates array with 10 values, going from 0 to 10 in 11 steps
y = x**2
print("x: ", x)
print("y: ", y)
plt.plot(x, y) # continuous line
```

```python
plt.scatter(x, y) # dots
```

```python
# c : color. Hex or name
# lw : line width
# ls : line style
plt.plot(x, y, c="#ff0033", lw=3, ls="dotted")
```

##### Create a graph from scratch

```python


plt.plot(x, y, label="square")
plt.plot(x ** 3, label="cube")
plt.title("Graph A")
plt.xlabel("base value")
plt.legend()
```

##### Save graph into file

```python
plt.savefig("graph.png")
```

##### Create a multi-graph figure

```python
# (height, width, select graph 'n')
plt.figure(figsize=(18,6))
plt.subplot(2,3,1)
plt.plot(x, y)
plt.subplot(2,3,2)
plt.plot(x ** 3)
plt.subplot(2,3,3)
plt.plot(x ** 4)
plt.subplot(2,3,4)
plt.plot(x / 2)
plt.subplot(2,3,5)
plt.plot(x / 3)
plt.subplot(2,3,6)
plt.plot(x / 4)
```

##### Graph types

###### Build dataset

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)
print(f"iris.data contains {x.shape[0]} examples and {x.shape[1]} variables")
print(f"There are {np.unique(y).size} unique classes")
```

###### Draw

```python
# column 0 & 1 from dataset
# color is determined by class, stored in y
# s : size of each point
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.8, s=x[:,2]*50)
plt.xlabel("sepal width")
plt.ylabel("sepal weight")
```
